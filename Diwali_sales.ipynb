```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Diwali_Sales_Data.csv",encoding='unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
print(df)
```

```python
df.head(10)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype("int")
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax=sns.countplot(x='Gender',data=df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

AGE:-


```python
ax=sns.countplot(data=df,x="Age Group",hue='Gender')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.columns
```

```python
sales_gen=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_gen)
```

```python
sales_State=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by='Orders',ascending=False).head(10)
# sns.set_theme(rc={'figure.figsize':(25,5)})
# plt.figure(figsize=(16,9))

ax=sns.barplot(x='State',y='Orders',data=sales_State)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
sales_amt=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False).head(10)
# sns.set_theme(rc={'figure.figsize':(16,9)})
# plt.figure(figsize=(16,9))

ax=sns.barplot(x='State',y='Amount',data=sales_amt)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
ax=sns.countplot(data=df,x="Marital_Status")
# plt.figure("figsize=(5,5)")
# sns.set_theme(rc={'figure.figsize':(16,9)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_State=df.groupby(["Marital_Status",'Gender'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
# sns.set_theme(rc={'figure.figsize':(25,5)})
# plt.figure(figsize=(16,9))

sns.barplot(x='Marital_Status',y='Amount',data=sales_State,hue="Gender")
```

```python
ax=sns.countplot(data=df,x="Occupation")
# plt.figure("figsize=(5,5)")
# sns.set_theme(rc={'figure.figsize':(10,9)})
for bar in ax.containers:
    ax.bar_label(bar)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
sales_State=df.groupby(['Occupation'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
# sns.set_theme(rc={'figure.figsize':(25,5)})
plt.figure(figsize=(16,9))

ax=sns.barplot(x='Occupation',y='Amount',data=sales_State)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```
